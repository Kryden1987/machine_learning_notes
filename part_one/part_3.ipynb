# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import networkx as nx

fig = plt.figure()  

x = np.random.rand(2,5)

print(x)

plt.scatter(x[0],x[1])

plt.show()
```

```python
diap = np.arange(0,30,0.1)
sn   = np.sin(diap)

plt.plot(diap,sn)
plt.show()
```

```python
 mnist = tf.keras.datasets.mnist
(X_train, y_train),(X_test, y_test) = mnist.load_data()

image_ten = np.array(X_train[10], dtype='float')
pixels = image_ten.reshape((28, 28))
plt.imshow(pixels)
plt.show()
```

```python
G = nx.Graph()

G.add_edge('1','2')
G.add_edge('2','3')
G.add_edge('2','4')
G.add_edge('3','5')
G.add_edge('3','6')
G.add_edge('4','7')
G.add_edge('4','8')
G.add_edge('4','9')

pos = nx.spring_layout(G)

nx.draw_networkx_labels(G, pos)

nx.draw(G, pos, font_size=16, with_labels=False)

plt.show()
```
