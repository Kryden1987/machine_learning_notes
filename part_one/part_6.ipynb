```python
import matplotlib.pyplot as plt
import matplotlib.image  as mpimg
import tensorflow        as tf
import skimage.measure
import numpy             as np
```

```python
img = mpimg.imread('../images/image.jpeg')

imgplot = plt.imshow(img)
```

```python
img_tensor = tf.placeholder(tf.float32, shape=[1,299,299,3])
pooled = tf.nn.max_pool(img_tensor, ksize=[1,16,16,1], strides=[1,16,16,1],padding='VALID')

img = img/255.0

with tf.Session() as sess:
    pooled_img = pooled.eval(feed_dict={img_tensor: img.reshape(1,299,299,3)})
    plt.imshow(np.squeeze(pooled_img, axis=0))
    
    print(pooled.shape)
```
