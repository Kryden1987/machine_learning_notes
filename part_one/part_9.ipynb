# Аугментация

```python
from   tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
image = plt.imread("../images/image.jpeg")
plt.imshow(image)
```

```python
datagen = ImageDataGenerator(rotation_range=50,width_shift_range=0.5,horizontal_flip=True)

augmented = datagen.random_transform(image)

plt.imshow(augmented)
```
